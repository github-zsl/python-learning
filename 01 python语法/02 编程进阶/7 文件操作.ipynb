文件操作问题和把大象放进冰箱问题类似。其核心思想为:**打开文件-->操作文件-->关闭文件**

## 1. 打开和关闭文件

打开文件使用的是open函数。<br>
&ensp;&ensp;open函数的基本语法如下：open(file_name [, access_mode][, buffering])<br>
【参数解析】<br>
- file_name变量：是一个包含要访问的文件名称的字符串值。
- access_mode变量：指打开文件的模式，对应有只读、写入、追加等。
- buffering：如果buffering的值被设为0，就不会有寄存；如果buffering的值取1，访问文件时就会寄存行；如果将buffering的值设为大于1的整数，表示这就是寄存区的缓冲大小；如果取负值，寄存区的缓冲大小就是系统默认的值。<br>

关闭文件使用close函数。

打开文件的模式:


```python
path = "C:/Users/ASUS/Desktop/pythonfiletest/firstfile.txt"
f = open(path)
print(f.name)
f.close()
```

## 3. csv 文件

## 5. 实际应用

```python
import os 
import time 

def batch_rename(path):
    global img_num  #定义全局变量
    if not os.path.isdir(path) and not os.path.isfile(path):
        return False
    if os.path.isfile(path):
        # 分割出目录与文件
        file_path = os.path.split(path)
        #分割出文件和文件扩展名
        lists = file_path[1].split('.')
        #取出后缀名（列表的切片操作）
        file_ext = lists[-1]
        img_ext = ['bmp', 'jpeg', 'png', 'gif', 'psd', 'jpg']
        if file_ext in img_ext:
            os.rename(path, file_path[0] + '/' + str(img_num) + '.png')
            #os.rename(path, file_path[0] + '/' + str(img_num) + '.'+file_ext)
            img_num += 1
            
    elif os.path.isdir(path):
        for item in os.listdir(path):
            #递归调用
            batch_rename(os.path.join(path,item))
            
if __name__ == "__main__":
    img_dir = "C:\\Users\\ASUS\\Desktop\\pythonfiletest\\images"
    path = img_dir.replace('\\','/')
    start  = time.time()
    img_num = 1
    batch_rename(path)
    print('总共处理了 %s 张图片,耗时:%0.2f.' %(img_num-1, time.time() - start))
```
