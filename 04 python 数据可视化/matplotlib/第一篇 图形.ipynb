## 第一章 向几何图形里填充颜色

### 1. 多边形的颜色填充

#### 1.1 规则多边形的颜色填充

```python
import matplotlib.pyplot as plt
import numpy as np

x= [0,0,5,10,15,15,10,5]
y= [5,10,15,15,10,5,0,0]

plt.fill(x,y,color="cornflowerblue")  #fill函数完成绘制八边形 xy表示封闭区域顶点的有序数对
plt.xlim(-1,16)  #完成多边形相对位置的调整
plt.ylim(-1,16)

plt.xticks(np.arange(0,16,5)) #调整刻度线的显示位置
plt.yticks(np.arange(0,16,5))
plt.show()
```

#### 1.2 不规则多边形的颜色填充

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi,500)
y = np.sin(x)

plt.fill(x,y,color="cornflowerblue",alpha=0.5)
#通过fill()函数绘制正弦函数
plt.plot(x,y,color="cornflowerblue",alpha=2.0)
#通过折线函数plot给填充区域添加轮廓 alpha参数设定颜色透明度
plt.plot([x[0],x[-1]],[y[0],y[-1]],color="cornflowerblue",alpha=0.8)


plt.xlim(0,2*np.pi)
plt.ylim(-1,1,1.1)
plt.show()
```

### 2.交叉曲线的颜色填充

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,500)
y1 = np.sin(2 * np.pi * x)
y2 = 1 * np.sin(3 * np.pi * x)

fig, ax = plt.subplots(3,1,sharex="all") #3行1列的共享x轴的网格布局的子区。

# "between y2 and y=0"
ax[0].fill_between(x,0,y2,alpha=0.5)
#调用实例方法fill_between（）填充由曲线y2和曲线y=0交叉围成的公共区域的颜色，同时，使用参数alpha调整填充区域的透明度。
ax[0].set_ylim(-1.2,1.2)


# "between y2 and 1 "
ax[1].fill_between(x,1,y2,alpha=0.5)
ax[1].set_ylim(-1.2,1.2)


# "between y1 and y2"
ax[2].fill_between(x,y1,y2,alpha=0.5)
ax[2].set_xlim(0,2)
ax[2].set_ylim(-1.2,1.2)

plt.show()
```
